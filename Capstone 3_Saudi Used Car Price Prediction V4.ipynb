Selanjutnya adalah melakukan tes distribusi, 
Mengapa Perlu Melakukan Tes Distribusi Normal dan Menggunakan Shapiro-Wilk Test?

- **Memahami Karakteristik Data:**  
  Tes distribusi normal membantu kita memahami apakah data mengikuti pola distribusi normal (Gaussian). Ini penting karena banyak analisis statistik dan model prediktif mengasumsikan data atau residual yang berdistribusi normal.

- **Validasi Asumsi Statistik:**  
  Beberapa metode statistik (misalnya, uji parametrik atau analisis regresi) mengandalkan asumsi normalitas. Jika data tidak normal, hasil uji statistik mungkin tidak valid atau interpretasinya bisa menyesatkan.

- **Pemilihan Teknik Pemodelan:**  
  Mengetahui distribusi data memungkinkan kita menentukan apakah perlu dilakukan transformasi data (seperti log-transformasi) untuk mencapai normalitas atau memilih model yang tidak terlalu sensitif terhadap distribusi data.

- **Mengapa Shapiro-Wilk Test?**  
  - **Sensitivitas:** Shapiro-Wilk adalah salah satu tes yang paling sensitif untuk mendeteksi penyimpangan dari distribusi normal, terutama pada sampel berukuran kecil hingga sedang.
  - **Kemudahan Implementasi:** Tes ini mudah diimplementasikan dengan library seperti SciPy dan memberikan p-value yang membantu dalam pengambilan keputusan (misalnya, p-value < 0.05 menunjukkan data kemungkinan tidak berdistribusi normal).

Dengan melakukan tes distribusi normal, kita dapat memastikan bahwa asumsi-asumsi penting terpenuhi atau mengetahui kapan harus melakukan penyesuaian agar analisis atau model yang dibangun lebih akurat dan valid.


```python
from scipy.stats import shapiro
import warnings
warnings.simplefilter("ignore")

numeric_cols = ['Price', 'Mileage', 'Year', 'Engine_Size']

for col in numeric_cols:
    print("="*60)
    print(f"Analisis Kolom: {col}")
    
    # 1. Histogram + KDE
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribusi Kolom {col}')
    plt.show()
    
    # 2. Shapiro-Wilk Test
    # Pastikan kolom tidak mengandung NaN sebelum dijalankan
    data_non_null = df[col].dropna()
    stat, p_val = shapiro(data_non_null)
    
    print(f"Shapiro-Wilk Test untuk kolom {col}:")
    print(f"  Statistik = {stat:.4f}, p-value = {p_val:.4f}")
    if p_val < 0.05:
        print("  => Data kemungkinan TIDAK terdistribusi normal (Reject H0).")
    else:
        print("  => Data berpotensi terdistribusi normal (Fail to reject H0).")
    print("="*60)
    print("\n")
```

**Important Note** : Karena datanya tidak terdistribusi normal maka kita akan menggunakan metode IQR untuk mendeteksi outliers.

### Menggunakan Metode IQR untuk Deteksi Outliers

- **Tidak Bergantung pada Asumsi Normalitas:**  
  IQR merupakan metode non-parametrik yang tidak mengasumsikan data berdistribusi normal, sehingga cocok untuk dataset yang tidak terdistribusi normal.

- **Robust terhadap Outliers:**  
  Dengan memfokuskan analisis pada rentang kuartil (Q1 sampai Q3), IQR mengabaikan nilai ekstrem sehingga lebih tahan terhadap pengaruh outlier dibandingkan dengan metode berbasis mean dan standar deviasi.

- **Simplicity dan Efektivitas:**  
  IQR mudah dihitung dan diinterpretasikan dengan menggunakan aturan 1.5 * IQR untuk menentukan batas bawah dan atas, membuatnya menjadi pendekatan yang efektif untuk mendeteksi nilai-nilai yang menyimpang secara signifikan.

- **Fokus pada Data Tengah:**  
  Metode ini menangkap informasi dari 50% data tengah (antara Q1 dan Q3), yang memberikan gambaran yang lebih stabil tentang sebaran data dan membantu mengidentifikasi outliers dengan lebih tepat.


```python
numeric_columns = ['Year', 'Engine_Size', 'Mileage', 'Price']

for col in numeric_columns:
    # 1. Menghitung Q1, Q3, dan IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # 2. Menentukan batas bawah dan batas atas
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # 3. Filter baris-baris yang termasuk outlier
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    # 4. Menampilkan jumlah outlier dan persentase relatif terhadap total baris
    total_rows = len(df)
    num_outliers = len(outliers)
    percentage_outliers = (num_outliers / total_rows) * 100

    print(f"Kolom: {col}")
    print(f"Jumlah Outlier: {num_outliers} dari {total_rows} baris data ({percentage_outliers:.2f}%).")
    print(f"Batas Bawah = {lower_bound}, Batas Atas = {upper_bound}")
    
    if num_outliers > 0:
        # Menampilkan semua nilai outlier yang unik untuk kolom tersebut
        unique_outliers = sorted(outliers[col].unique())
        print(f"Nilai Outlier Unik untuk '{col}': {unique_outliers}")
    
    print("-"*50)
    
    # 5. Visualisasi dengan Boxplot
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    
    # Tambahkan garis bantu untuk lower_bound dan upper_bound (opsional)
    plt.axvline(x=lower_bound, color='red', linestyle='--', label='Lower Bound')
    plt.axvline(x=upper_bound, color='blue', linestyle='--', label='Upper Bound')
    
    plt.title(f'Boxplot untuk Kolom {col}')
    plt.legend()
    plt.show()
```

**Important Point** : Berdasarkan apa yang sudah kita deteksi mengenai outliers, maka kita akan mengambil langkah dengan menghapus outliers dengan nilai yang ekstrem saja. Berikut yang akan kita lakukan :
1. Kolom year : Hapus tahun dibawah 1970, yaitu 1963, 1964, dan 1965.

```python
# Simpan jumlah baris sebelum pembersihan
initial_count = len(df)

# 1. Hapus baris dengan kolom 'Year' < 1970
df = df[df['Year'] >= 1970]

# 2. Engine_Size dipertahankan (tidak dihapus outliernya),
#    sehingga tidak ada filter untuk 'Engine_Size'.

# 3. Hapus baris dengan kolom 'Mileage' > 5.000.000
df = df[df['Mileage'] <= 5000000]

# 4. Hapus baris dengan kolom 'Price' > 650.000
df = df[df['Price'] <= 650000]

# Simpan jumlah baris setelah pembersihan
final_count = len(df)
removed_rows = initial_count - final_count

print(f"Jumlah baris awal: {initial_count}")
print(f"Jumlah baris setelah menghapus outliers: {final_count}")
print(f"Total baris yang dihapus: {removed_rows}")
```

5. Melakukan pengecekan korelasi dengan metode spearman karena data kita tidak terdistribusi normal

- **Mengapa Menggunakan Metode Spearman:**  
  - Spearman adalah metode non-parametrik yang mengukur korelasi berdasarkan peringkat (ranks), sehingga tidak terpengaruh oleh asumsi normalitas data.
  - Metode ini mampu mendeteksi hubungan monotonic antara variabel, meskipun hubungan tersebut tidak linier, sehingga lebih cocok untuk dataset yang tidak terdistribusi normal.

- **Mengapa Melihat Korelasi Antar Fitur Numerikal:**  
  - Analisis korelasi membantu mengidentifikasi hubungan antara fitur, sehingga kita bisa memahami apakah terdapat fitur yang sangat berkorelasi dan berpotensi redundan.
  - Mengetahui korelasi antar fitur juga berguna untuk mendeteksi multikolinearitas, yang dapat mempengaruhi stabilitas dan interpretabilitas model prediktif.


```python
# Daftar kolom yang ingin dilihat korelasinya
cols = ['Year', 'Engine_Size', 'Mileage', 'Price']

# 1) Membuat subset DataFrame hanya dengan 4 kolom numerik tersebut
df_corr = df[cols].copy()

# 2) Menghitung korelasi dengan metode Spearman (karena data non-normal)
corr_matrix = df_corr.corr(method='spearman')

# 3) Menampilkan nilai korelasi dalam bentuk tabel
print("Matriks Korelasi (Spearman):")
print(corr_matrix)

# 4) Visualisasi dengan Heatmap Seaborn
plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, 
            annot=True,        # menampilkan nilai korelasi di tiap sel
            cmap='coolwarm',   # skema warna
            vmin=-1, vmax=1,   # mengatur rentang warna dari -1 s/d 1
            square=True)       # sel heatmap berbentuk kotak

plt.title('Spearman Correlation Heatmap')
```

- **Binary Encoding:**  
  - Mengubah variabel kategorikal dengan banyak kategori menjadi representasi biner, sehingga mengurangi dimensi fitur dan cocok untuk fitur dengan cardinality tinggi.

- **One Hot Encoding:**  
  - Mengonversi variabel kategorikal ke dalam format numerik dengan membuat kolom terpisah untuk setiap kategori, sehingga mencegah model mengasumsikan adanya urutan antar kategori.

- **Standard Scaler:**  
  - Menstandarkan fitur numerik agar memiliki mean 0 dan standar deviasi 1, sehingga membantu model machine learning dalam konvergensi dan menghindari bias yang diakibatkan oleh skala fitur yang berbeda.


Type, Region, Make akan menggunakan binary encoding karena ada banyak unique values dan tidak memiliki urutan.

### Penjelasan Detail Cara Kerja Algoritma Model dan Kriteria Kepercayaan Model

### Cara Kerja Algoritma Model

### Base Models
- **Linear Regression:**  
  Mengestimasikan hubungan linier antara fitur dan target dengan mencari koefisien yang meminimalkan error kuadrat, ideal untuk hubungan yang linear dan data bebas multikolinearitas serta outlier ekstrim.

- **KNN Regressor:**  
  Menghitung nilai target berdasarkan rata-rata nilai dari 'k' tetangga terdekat dalam ruang fitur; efektif untuk pola lokal, namun sensitif terhadap skala fitur dan data berdimensi tinggi.

- **Decision Tree Regressor:**  
  Membagi data secara rekursif dengan membuat keputusan berdasarkan nilai fitur, sehingga menghasilkan struktur pohon yang memetakan fitur ke target; mampu menangkap hubungan non-linier namun rentan terhadap overfitting jika tidak dikendalikan.

- **Ridge, Lasso, dan Elastic Net:**  
  Merupakan varian Linear Regression dengan penambahan regularisasi (L2 untuk Ridge, L1 untuk Lasso, dan kombinasi keduanya untuk Elastic Net) yang membantu mengatasi multikolinearitas dan mencegah overfitting dengan mengurangi kompleksitas model.

### Ensemble Models
- **Random Forest Regressor:**  
  Menggabungkan prediksi dari banyak decision tree yang dilatih secara acak pada subset data dan fitur (bagging) untuk mengurangi variansi dan meningkatkan stabilitas prediksi.

- **AdaBoost Regressor:**  
  Mengadaptasi bobot data secara iteratif, sehingga model lebih fokus pada contoh yang sulit diprediksi; membantu meningkatkan akurasi dengan mengurangi bias, namun bisa terpengaruh oleh outlier.

- **Gradient Boosting Regressor:**  
  Membangun model secara berurutan dengan setiap pohon mencoba memperbaiki kesalahan residual dari model sebelumnya; efektif menangkap pola kompleks, namun memerlukan tuning agar tidak overfit.

- **Xtreme Gradient Boosting (XGBoost):**  
  Versi yang dioptimasi dari gradient boosting yang menambahkan regularisasi dan teknik komputasi efisien, sehingga cocok untuk dataset besar dan kompleks dengan kinerja tinggi bila telah dituning dengan baik.

### Kapan Model ML Dapat Dipercaya dan Kapan Tidak

### Model Dapat Dipercaya (Akurat) Jika:
- **Data Training Representatif:**  
  Data yang digunakan untuk melatih model mencakup seluruh variabilitas yang mungkin ditemui di dunia nyata, sehingga model dapat menangani kasus baru dengan karakteristik serupa.

- **Asumsi Model Terpenuhi:**  
  Misalnya, untuk Linear Regression, hubungan antara fitur dan target bersifat linier; untuk model tree-based, data cukup bersih dari noise ekstrim.

- **Validasi dan Cross-Validation Konsisten:**  
  Metrik performa (seperti RMSE, MAE, dan R²) menunjukkan performa yang baik dan konsisten antara data pelatihan, validasi, dan test set, serta analisis residual menunjukkan error yang tersebar acak.

- **Data Baru Sesuai dengan Data Training:**  
  Input baru yang akan diprediksi memiliki karakteristik (rentang nilai, distribusi) yang mirip dengan data yang digunakan untuk melatih model.

### Model Tidak Dapat Dipercaya (Kurang Akurat) Jika:
- **Data Tidak Representatif atau Out-of-Distribution:**  
  Data baru yang memiliki pola, rentang, atau distribusi yang sangat berbeda dari data training, seperti data dengan outlier ekstrim atau kondisi yang belum pernah terjadi selama pelatihan.

- **Overfitting atau Underfitting Terjadi:**  
  Model yang terlalu kompleks mungkin menghafal noise pada data pelatihan (overfitting) atau model yang terlalu sederhana tidak menangkap pola penting (underfitting), sehingga performa di dunia nyata menurun.

- **Data Bermasalah:**  
  Data yang banyak mengandung noise, missing values, atau kesalahan pencatatan dapat mengurangi akurasi prediksi model.

### Data yang Membuat Model Bisa Dipercaya dan Tidak dipercaya

- **Data yang Membuat Model Bisa Dipercaya (Akurat):**
  - Data yang sudah melalui proses pembersihan dan preprocessing (misalnya, imputasi missing values, scaling, encoding) dengan konsisten.
  - Data yang mencakup rentang nilai dan distribusi yang sama dengan data training.
  - Data dengan sedikit noise dan outlier yang sudah diatasi melalui teknik seperti IQR.

- **Data yang Membuat Model Tidak Dapat Dipercaya (Kurang Akurat):**
  - Data yang memiliki fitur atau pola yang berbeda secara signifikan dibandingkan dengan data training (misalnya, perubahan tren pasar yang belum tercermin di data lama).
  - Data yang mengandung banyak outlier, noise, atau missing values yang tidak ditangani dengan baik.
  - Data dari segmen yang jarang atau tidak terwakili dalam data training, sehingga model kesulitan memberikan prediksi yang akurat.

## Evaluasi Model Menggunakan RMSE, MAE, MAPE, dan R²

Dalam proyek prediksi harga mobil bekas, digunakan **4 metrik evaluasi** untuk menilai performa model, yaitu RMSE, MAE, MAPE, dan R². Di bawah ini dijelaskan cara kerja masing-masing metrik dan bagaimana kaitannya dengan aspek bisnis:

## 1. Root Mean Squared Error (RMSE)
- **Definisi:**  
  RMSE dihitung sebagai akar dari rata-rata kuadrat selisih antara nilai prediksi dan nilai aktual, yaitu:  
  $$
RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
$$
- **Cara Kerja:**  
  Dengan mengkuadratkan selisih, RMSE memberikan penalti lebih besar terhadap error yang besar; kemudian, dengan mengakarnya, hasilnya dikembalikan ke satuan yang sama dengan target.
- **Kelebihan:**  
  - Sensitif terhadap error besar, sehingga mengungkapkan jika ada prediksi yang jauh meleset.  
  - Satuan output sama dengan target (misalnya, rupiah), memudahkan interpretasi.
- **Hubungan dengan Aspek Bisnis:**  
  Dalam konteks bisnis, kesalahan besar dalam prediksi harga dapat mengakibatkan keputusan yang merugikan (misalnya, overpricing atau underpricing mobil). RMSE membantu mengidentifikasi dan meminimalkan dampak kesalahan ekstrim tersebut sehingga keputusan penetapan harga lebih andal.

## 2. Mean Absolute Error (MAE)
- **Definisi:**  
  MAE adalah rata-rata selisih absolut antara prediksi dan nilai aktual:  
  $$
MAE = \frac{1}{n} \sum_{i=1}^{n} \left| y_i - \hat{y}_i \right|
$$
- **Cara Kerja:**  
  MAE menghitung nilai kesalahan tanpa memperhatikan arah (positif atau negatif) dan memberikan gambaran rata-rata seberapa jauh prediksi menyimpang dari nilai sebenarnya.
- **Kelebihan:**  
  - Mudah diinterpretasikan karena memberikan nilai error dalam satuan yang sama.  
  - Kurang sensitif terhadap outlier dibandingkan RMSE.
- **Hubungan dengan Aspek Bisnis:**  
  MAE memberikan pandangan yang realistis tentang seberapa besar rata-rata selisih prediksi dengan harga sebenarnya. Ini sangat berguna dalam penganggaran dan perencanaan keuangan karena menyampaikan perkiraan kerugian atau keuntungan secara rata-rata dalam transaksi harga mobil.

## 3. Mean Absolute Percentage Error (MAPE)
- **Definisi:**  
  MAPE mengukur kesalahan rata-rata dalam bentuk persentase dari nilai aktual:  
  $$
MAPE = \frac{100\%}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right|
$$
- **Cara Kerja:**  
  Dengan menghitung error relatif terhadap nilai aktual, MAPE menghasilkan nilai persentase yang menunjukkan seberapa besar kesalahan dibandingkan dengan nilai yang diharapkan.
- **Kelebihan:**  
  - Hasil berupa persentase memudahkan pemahaman oleh pemangku kepentingan non-teknis.  
  - Memungkinkan perbandingan antar model dan kondisi pasar yang berbeda.
- **Hubungan dengan Aspek Bisnis:**  
  Dalam bisnis, MAPE memudahkan analisis risiko dan profitabilitas karena menunjukkan error dalam konteks persentase. Misalnya, mengetahui bahwa model memiliki error 5% dapat langsung diterjemahkan ke dalam estimasi keakuratan harga dan potensi margin keuntungan atau kerugian.

## 4. R-squared (R²)
- **Definisi:**  
  R² mengukur proporsi variabilitas target yang dapat dijelaskan oleh model, dihitung dengan:  
  $$
R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}
$$

- **Cara Kerja:**  
  R² membandingkan total varians yang dijelaskan oleh model dengan total varians dalam data aktual. Nilai mendekati 1 berarti model dapat menjelaskan sebagian besar variabilitas.
- **Kelebihan:**  
  - Memberikan indikasi seberapa baik model menangkap pola dan hubungan dalam data.  
  - Mudah dipahami, karena nilainya berkisar antara 0 (tidak ada penjelasan) hingga 1 (penjelasan sempurna).
- **Hubungan dengan Aspek Bisnis:**  
  R² memberikan gambaran kepercayaan terhadap model dalam menjelaskan harga mobil. Nilai yang tinggi menunjukkan bahwa model dapat diandalkan dalam membuat prediksi, sehingga strategi bisnis, seperti penetapan harga atau evaluasi pasar, dapat dilakukan dengan keyakinan lebih tinggi.

## HyperParameter Tuning

Langkah selanjutnya adalah melakukan Hyperparameter Tuning terhadap model yang kita pilih. Metode Hyperparameter Tuning yang dipilih adalah dengan Random Search, tujuannya adalah agar pencarian bisa menjadi lebih cepat dibandingkan jika kita menggunakan Grid Search.

- **Mengapa Menggunakan Random Search:**
  - RandomizedSearchCV memilih kombinasi parameter secara acak dari ruang yang telah ditentukan, sehingga lebih efisien daripada grid search ketika ruang parameter cukup besar.
  - Dengan `n_iter=50`, kita mengeksplorasi 50 kombinasi secara acak, yang memberikan trade-off antara cakupan ruang parameter dan waktu komputasi.

- **Alasan Pemilihan Parameter:**
  - **n_estimators :** Menentukan jumlah pohon dalam model; rentang ini membantu menemukan jumlah pohon optimal untuk menghindari underfitting atau overfitting.
  - **max_depth :** Mengontrol kedalaman maksimum setiap pohon; membantu mengatur kompleksitas model dan menghindari overfitting.
  - **learning_rate :** Menentukan seberapa besar setiap pembaruan model; nilai yang lebih kecil bisa menghasilkan model yang lebih stabil, meskipun membutuhkan lebih banyak pohon.
  - **subsample :** Mengatur persentase sampel data yang digunakan untuk membangun setiap pohon; dapat membantu mengurangi overfitting dengan melakukan sampling acak.
  - **colsample_bytree :** Menentukan persentase fitur yang digunakan untuk setiap pohon; juga membantu meningkatkan generalisasi model.
  - **gamma :** Menetapkan minimum loss reduction yang diperlukan untuk melakukan split; nilai ini membuat model lebih konservatif dalam membentuk pohon.
  - **min_child_weight :** Mengontrol jumlah minimum instance yang dibutuhkan dalam sebuah leaf; membantu menghindari split yang terlalu kecil dan overfitting.

- **Setup RandomizedSearchCV:**
  - **cv=5:** Menggunakan 5-fold cross-validation untuk mendapatkan estimasi performa model yang stabil dan menghindari overfitting.
  - **scoring='r2':** R² dipilih sebagai metrik evaluasi karena memberikan gambaran seberapa baik model dapat menjelaskan variabilitas data.
  - **n_iter=50:** Menjalankan 50 iterasi pencarian memungkinkan eksplorasi ruang parameter yang luas tanpa harus mencoba semua kombinasi.
  - **n_jobs=-1:** Memanfaatkan semua core CPU yang tersedia untuk mempercepat proses pencarian.
  - **random_state=42:** Menjamin reprodusibilitas hasil pencarian kombinasi parameter.

Secara keseluruhan, setup ini dirancang untuk mengoptimalkan Xtreme Gradient Boosting Regressor secara efisien, sehingga dapat meningkatkan performa prediksi harga mobil bekas dengan memanfaatkan kombinasi parameter terbaik yang ditemukan melalui RandomizedSearchCV.


```python
from sklearn.model_selection import RandomizedSearchCV

# Definisikan distribusi parameter yang ingin ditelusuri
param_dist = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [3, 4, 5, 6, 7],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0],
    'gamma' : [0, 0.1, 0.2, 0.3],
    'min_child_weight' : [1, 3, 5]
}

# Inisialisasi model XGBRegressor dengan parameter dasar
xgb_model = XGBRegressor(objective="reg:squarederror", random_state=42)

# Setup RandomizedSearchCV dengan 5-fold cross validation
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=50,                # jumlah iterasi pencarian
    scoring='r2',  # menggunakan r2 sebagai scoring
    cv=5,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# Lakukan fitting pada data training
random_search.fit(X_train, y_train)

# Cetak parameter terbaik dan skor terbaik
print("Best parameters found:", random_search.best_params_)
print("Best score R2:", random_search.best_score_)

# Ambil model terbaik dari hasil pencarian
best_xgb = random_search.best_estimator_

# Lakukan prediksi pada data testing menggunakan model terbaik
y_pred = best_xgb.predict(X_test)

# Evaluasi model menggunakan R2, RMSE, MAE, dan MAPE
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)

# Buat DataFrame untuk menampilkan hasil evaluasi secara rapi
results_df = pd.DataFrame({
    'RMSE': [rmse],
    'MAE': [mae],
    'MAPE': [mape],
    'R2': [r2]
}, index=['Tuned XGBoost Regressor (Random Search)'])

results_df
```

## Feature Importance

Berikut ini adalah grafik yang menunjukkan feature importance, sehingga kita bisa mengetahui kolom mana saja yang paling berkontribusi dalam hasil prediksi model. disini kita menggunakan beberapa jenis importance_type, yaitu : 

1. **weight**  
   - Menghitung seberapa sering (berapa kali) sebuah fitur digunakan untuk melakukan split di seluruh pohon.  
   - Semakin tinggi nilai ini, semakin sering fitur tersebut dianggap relevan untuk pemisahan data.

2. **gain**  
   - Mengukur rata-rata peningkatan (gain) yang diperoleh setiap kali fitur tersebut digunakan dalam split.  
   - Semakin tinggi nilainya, semakin besar kontribusi fitur terhadap pengurangan loss.

3. **cover**  
   - Mengukur rata-rata banyaknya data (instance) yang terpengaruh oleh split menggunakan fitur tersebut.  
   - Semakin tinggi nilainya, semakin banyak data yang “dicakup” oleh split fitur tersebut.

Pilihan `importance_type` mana yang digunakan tergantung kebutuhan analisis :  
- *weight* lebih sederhana karena hanya menghitung frekuensi penggunaan fitur,  
- *gain* dan *cover* memberikan gambaran seberapa besar manfaat fitur tersebut,  

```python
from xgboost import plot_importance

# Pastikan loaded_model adalah model XGBoost yang sudah diload atau model terbaik hasil tuning
plt.figure(figsize=(10, 8))
plot_importance(best_xgb, max_num_features=20, importance_type='gain')  # max_num_features opsional untuk membatasi jumlah fitur yang ditampilkan
plt.title("Feature Importance - XGBoost Model")
plt.show()
```

```python
plot_importance(best_xgb, max_num_features=20, importance_type='weight')  # max_num_features opsional untuk membatasi jumlah fitur yang ditampilkan
plt.title("Feature Importance - XGBoost Model")
plt.show()
```

```python
plot_importance(best_xgb, max_num_features=20, importance_type='cover')  # max_num_features opsional untuk membatasi jumlah fitur yang ditampilkan
plt.title("Feature Importance - XGBoost Model")
plt.show()
```

Dari feature importance di atas, dapat diketahui bahwa fitur Make, Year, dan Engine_Size sangat memengaruhi model. karena ketiganya memiliki frekuensi kemunculan yang tinggi dalam trees dan memberikan peningkatan akurasi ketika fitur digunakan untuk split.

### Analisis Feature Importance Berdasarkan Tipe (Gain, Weight, dan Cover)

Dalam XGBoost, **feature importance** dapat diukur dengan berbagai cara tergantung pada parameter `importance_type`. Meskipun tujuannya sama (mengetahui fitur yang paling berpengaruh), ketiga tipe ini menekankan aspek yang berbeda:

1. **Gain**  
   - **Interpretasi Hasil:**  
     - Fitur dengan **gain** tinggi berkontribusi besar dalam mengurangi error ketika digunakan dalam pemisahan (split).  
     - Sebagai contoh, jika satu fitur jarang dipakai tetapi setiap kali dipakai mampu mengurangi error secara signifikan, maka *gain*-nya bisa sangat tinggi.  
   - **Contoh Observasi:**  
     - Jika *Options_Full* muncul dengan nilai *gain* yang sangat besar, artinya meski mungkin tidak sering dipakai, setiap split yang melibatkan *Options_Full* memberikan penurunan error yang substansial.

2. **Weight**  
   - **Interpretasi Hasil:**  
     - Fitur dengan **weight** tinggi sering muncul dalam proses splitting, walaupun setiap split-nya belum tentu memberikan *gain* yang besar.  
     - Fitur seperti *Mileage* atau *Engine_Size* yang sering dipakai menandakan model sering membagi data berdasarkan fitur tersebut.  
   - **Contoh Observasi:**  
     - Jika *Mileage* memiliki *weight* tertinggi, berarti split berdasarkan *Mileage* paling sering terjadi di banyak node pohon, meski *gain* per split bisa bervariasi.

3. **Cover**  
   - **Interpretasi Hasil:**  
     - Fitur dengan **cover** tinggi cenderung membagi subset data yang cukup besar, sehingga memengaruhi banyak instance.  
     - Menunjukkan sejauh mana fitur tersebut *“mencakup”* data dalam setiap split.  
   - **Contoh Observasi:**  
     - Jika *Mileage* memiliki *cover* yang sangat besar, berarti saat *Mileage* digunakan untuk split, ia menangani kelompok data yang luas.

### Hasilnya
- **Gain** menitikberatkan pada besarnya kontribusi setiap split dalam menurunkan error.
- **Weight** menekankan frekuensi pemakaian fitur dalam proses pemisahan.
- **Cover** menunjukkan seberapa banyak data yang dipengaruhi oleh split tersebut.

Dalam praktiknya, ketiga tipe ini saling melengkapi:
- Fitur dengan *gain* tinggi **sangat** penting dalam menurunkan error meskipun frekuensi penggunaannya mungkin tidak banyak.
- Fitur dengan *weight* tinggi sering dipakai model, meski peningkatan performanya per split mungkin tidak sebesar fitur lain.
- Fitur dengan *cover* tinggi berdampak pada banyak instance setiap kali digunakan untuk split, yang juga mencerminkan pentingnya fitur dalam memisahkan data secara efektif.
